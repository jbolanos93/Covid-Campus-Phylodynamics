### Step 1: Install Libraries

```r
# Function to install missing packages
install_if_missing <- function(packages) {
    to_install <- packages[!(packages %in% installed.packages()[,"Package"])]
    if(length(to_install)) install.packages(to_install, dependencies = TRUE)
}

# List of required libraries
required_packages <- c("circlize", "glue", "ComplexHeatmap")

# Install missing packages
install_if_missing(required_packages)

# Load libraries
library(circlize)
library(glue)
library(ComplexHeatmap)
```

```r

ChordDiagramPlotGrouped_PDF_v2 <- function(a_csv, DiagramTitle) {
  library(circlize)
  library(glue)
  library(ComplexHeatmap)
  
  # Read data
   ChordData<-read.csv(a_csv)

  # Visualize data
    print(ChordData)
  # Set up color scheme
  grid.col = c(Off_Campus = "orange", On_Campus = "blue")
  
    #adjust plot size in notebook, ignore this line if not using jupyter notebook.
      options(repr.plot.width = 9, repr.plot.height = 8)
  # Create chord diagram
  chordDiagram(ChordData, grid.col = grid.col, annotationTrack = 'grid')
  # Create legend
  lgd_lines <- Legend(
    labels = c("Off Campus", "On Campus"),
    legend_gp = gpar(fill = c("darkorange", "blue")),
    title = "Transitions",
    title_gp = gpar(fontsize = 12, fontface = "bold"),
    labels_gp = gpar(fontsize = 10)
  )
  
  # Add legend and title
  draw(lgd_lines, x = unit(0.05, "npc"), y = unit(0.80, "npc"), just = c("left", "top"))
  
  #Removind tittel for final editing, uncomment if you wnat tittle. 
  #title(DiagramTitle, cex = 1, line = -1)
  title(DiagramTitle, cex = 1, line = -1)
  
}

### Usage
#Path to BEAST log file 
Data<-"/Users/juanbolanos/Desktop/Phylogenetisc_Paper/Dornburg-Review/Final-Data/scripts/Visuals-CirclePlot-R/input/AllDataGrouped_Output_TransitionCounts_circlPlot-2024-04-15.csv"
ChordDiagramPlotGrouped_PDF_v2(Data,"Complete SARS-CoV-2 Transition Patterns\nBetween On-Campus and Off-Campus Populations")
```

## Interactive Circle Plot

```r
library(chorddiag)
csv_data<-read.csv("/Users/juanbolanos/Desktop/Phylogenetisc_Paper/Dornburg-Review/Final-Data/scripts/Visuals-CirclePlot-R/input/AllDataGrouped_Output_TransitionCounts_circlPlot-2024-04-15.csv")

# Visualize data
print(csv_data)
# Define the locations
locations <- c("Off_Campus", "On_Campus")


# Create an empty matrix with row and column names
matrix_data <- matrix(0, nrow = length(locations), ncol = length(locations))
rownames(matrix_data) <- locations
colnames(matrix_data) <- locations

# Fill the matrix with values from the CSV data
for (i in 1:nrow(csv_data)) {
matrix_data[csv_data$from[i], csv_data$to[i]] <- csv_data$value[i]
}

allDataGroupColors <- c("red", "darkgreen")
#adjust plot size in notebook, ignore this line if not using jupyter notebook.
options(repr.plot.width = 9, repr.plot.height = 8)

chorddiag(matrix_data, groupColors = allDataGroupColors, groupnamePadding = 20)


```
